```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
link=" https://www.amazon.in/Test-TST_Exclusive1003-Exclusive-1003/product-reviews/B0844592FT/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
uClient = urlopen(link)
page_html = uClient.read()
uClient.close()#Always close them
```

```python
page_soup = soup(page_html,"html.parser")#it will fitch all the data in html format
page_soup#It is unstructured data
```

```python
container = page_soup.find_all("span",class_="a-profile-name")#It available in div tag
```

```python
len(container)
```

```python
container
```

```python
cust_name=[]
for i in range(0,len(container)):
    cust_name.append(container[i].get_text())
```

### Product review title

```python
title = page_soup.find_all("a",class_="review-title")#It available in div tag
```

```python
len(title)
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
```

```python
review_title
```

```python
review_title[:]=[titles.lstrip("\n") for titles in review_title]
```

```python
review_title
```

```python
review_title[:]=[titles.rstrip("\n") for titles in review_title]
review_title
```

```python
review_rating = page_soup.find_all("i",class_="review-rating")#It available in div tag
```

```python
review_rating
```

```python
review_rating.pop(1)
```

```python
rating=[]
for i in range(0,len(review_rating)):
    rating.append(review_rating[i].get_text())
```

### Product review content

```python
review_content = page_soup.find_all("span",{"data-hook":"review-body"})#It available in div tag
```

```python
review_content
```

```python
contents=[]
for i in range(0,len(review_content)):
    contents.append(review_content[i].get_text())
```

```python
contents
```

```python
len(contents)
```

```python
contents[:]=[reviews.lstrip("\n") for reviews in contents]
contents
```

```python
contents[:]=[reviews.rstrip("\n") for reviews in contents]
contents
```

### cust_name, rating, contents (These are my data variables)

```python
len(cust_name)
```

```python
df=pd.DataFrame({"cust_name":cust_name,"rating":rating,"review_contents":contents})
```
